# Courbe $k_r(S_w)$ écoulement concentrique en tube capillaire 
Soit un tube capillaire de rayon $R$ dont la portion cylindrique concentrique de rayon $r_{1}$ est occupée par le fluide non-mouillant $\left(\gamma\right)$, tandis que la portion annulaire complémentaire est occupée par le fluide mouillant $\left(\beta\right)$. Les écoulements sont supposés parallèles et de même sens. Les écoulements des deux phases sont de type Poiseuille cylindrique, le profil de vitesse s'écrit  

$$v_{\alpha}=\frac{1}{4\mu_{\alpha}}r^{2}\frac{\partial p_{\alpha}}{\partial z}+a_{\alpha}\mathrm{ln}(r)+b_{\alpha}\quad,\alpha=\beta,\gamma.$$  


Les conditions limites associées sont les suivantes  


$$v_{\beta}(R)=0,$$  

$$\frac{\partial v_{\gamma}}{\partial r}=0\quad,r=0$$  

$$v_{\gamma}(r_{1})=v_{\beta}(r_{1})$$  

$$\mu_{\beta}\frac{\partial v_{\beta}}{\partial r}=\mu_{\gamma}\frac{\partial v_{\gamma}}{\partial r}\quad,r=r_{1}.$$  

Après utilisation des CL pour obtenir les constantes d'intégration on obtient les profils de vitesse suivants

$$v_{\beta}=\frac{\left(r^{2}-R^{2}\right)}{4\mu_{\beta}}\frac{\partial p_{\beta}}{\partial z}+\frac{r_{1}^{2}}{2\mu_{\beta}}\left(\frac{\partial p_{\gamma}}{\partial z}-\frac{\partial p_{\beta}}{\partial z}\right)\mathrm{ln}\left(\frac{r}{R}\right),$$

$$v_{\gamma}=\frac{\left(r^{2}-r_{1}^{2}\right)}{4\mu_{\gamma}}\frac{\partial p_{\gamma}}{\partial z}+\frac{r_{1}^{2}}{2\mu_{\beta}}\left(\frac{\partial p_{\gamma}}{\partial z}-\frac{\partial p_{\beta}}{\partial z}\right)\mathrm{ln}\left(\frac{r_{1}}{R}\right)+\frac{\left(r_{1}^{2}-R^{2}\right)}{4\mu_{\beta}}\frac{\partial p_{\beta}}{\partial z}.$$

Les vitesses moyennes superficielles s'écrivent

$$V_{\gamma}=\frac{1}{\pi R^{2}}\int_{0}^{r_{1}}2\pi rv_{\gamma}d\mathrm{r},$$  

$$V_{\beta}=\frac{1}{\pi R^{2}}\int_{r_{1}}^{R}2\pi rv_{\beta}d\mathrm{r},$$ 

Ce qui permet d'obtenir


$$V_{\gamma}=-\frac{1}{\mu_{\gamma}}\frac{\partial p_{\gamma}}{\partial z}\underset{K_{\gamma\gamma}^{*}}{\underbrace{\left[\frac{S_{\gamma}^{2}R^{2}}{8}\left(1-2\frac{\mu_{\gamma}}{\mu_{\beta}}\mathrm{ln}\left(S_{\gamma}\right)\right)\right]}}-\frac{1}{\mu_{\beta}}\frac{\partial p_{\beta}}{\partial z}\underset{K_{\gamma\beta}^{*}}{\underbrace{\left[\frac{S_{\gamma}R^{2}}{4}\left(1-S_{\gamma+}S_{\gamma}\mathrm{ln}\left(S_{\gamma}\right)\right)\right]}},$$

$$V_{\beta}=-\frac{1}{\mu_{\gamma}}\frac{\partial p_{\gamma}}{\partial z}\underset{K_{\beta\gamma}^{*}}{\underbrace{\left[\frac{\mu_{\gamma}}{\mu_{\beta}}\frac{S_{\gamma}R^{2}}{4}\left(1-S_{\gamma}+S_{\gamma}\mathrm{ln}\left(S_{\gamma}\right)\right)\right]}}-\\-\frac{1}{\mu_{\beta}}\frac{\partial p_{\beta}}{\partial z}\underset{K_{\beta\beta}^{*}}{\underbrace{\left[\frac{R^{2}}{8}\left(\left(1-S_{\gamma}\right)\left(1-3S_{\gamma}\right)-2S_{\gamma}^{2}\mathbf{\mathrm{ln}}\left(S_{\gamma}\right)\right)\right]}}.$$

Les perméabilités relatives sont obtenues en normalisant les expressions obtenues par la perméabilité absolue d'un tube $R^2/8$.

```python
#Parameters

M = 1 #Viscosity ratio (non-wetting over wetting)

#Plot permeabilities

import numpy as np
import matplotlib.pyplot as plt
from math import pi

x = np.linspace(1e-4,1-1e-4)

kww = x*(1-3*(1-x))-2*(1-x)*(1-x)*np.log(1-x)
knn = (1-x)*(1-x)*(1-2*M*np.log(1-x))
kwn = M*2*(1-x)*(x+(1-x)*np.log(1-x))
knw = kwn/M

#plt.figure(figsize=(8, 6), dpi=100)
plt.title('Relative permeability vs. wetting fluid saturation')

plt.plot(x,kww,label=r'$k_{\beta \beta}$')
plt.plot(x,knn,label=r'$k_{\gamma \gamma}$')
plt.plot(x,kwn,label=r'$k_{\beta \gamma}$')
plt.plot(x,knw,label=r'$k_{\gamma \beta}$')

plt.xlabel('Wetting fluid saturation')
plt.ylabel('Relative permeability')
plt.legend()
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')

#plt.show()
```
